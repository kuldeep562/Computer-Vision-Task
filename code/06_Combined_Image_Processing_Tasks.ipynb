```python
#sharp image

K = np.array([[0,-1,0],
              [-1,5,-1],
              [0,-1,0]])
sharp_img = cv2.filter2D(img2,-1,K)
plt.imshow(sharp_img)
```

```python
# img transformation 

t = np.float32([[1,0,50],
               [0,1,50]])
t_img = cv2.warpAffine(img,t,(300,300))
plt.imshow(t_img)
```

```python
# Box plot using function

kernel_size = (5,5)

filter_app = cv2.boxFilter(img, -1 , kernel_size)

cv2.imshow("org img",img)
cv2.imshow("fil omg",filter_app)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 2D filter

kernel = np.ones((4,4),np.float32) / 9

filter_img = cv2.filter2D(img2 , -1 , kernel)

cv2.imshow("org img",img2)
cv2.imshow("fil img", filter_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
