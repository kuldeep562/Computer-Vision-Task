```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def convo(path,kernel):
    image = cv2.imread(path,0)
    
    plt.figure(figsize=(7,7))
    plt.subplot(321)
    plt.title("Original Image")
    plt.imshow(image,cmap='gray')
    
    plt.subplot(322)
    plt.title("Histogram of original image")
    plt.hist(image.flatten(), bins=256, range=[0,256], density=True)
    
    h,w = image.shape
    for i in range(h-2):
        for j in range(w-2):
            mat = np.array(image[i:i+3,j:j+3])
            ad = np.array([mat*kernel])
            ad = np.sum(ad)
            image[i,j] = abs(ad)

    plt.subplot(323)
    plt.title("convolved Image")
    plt.imshow(image,cmap='gray')
    
    plt.subplot(324)
    plt.title("Histogram of original image")
    plt.hist(image.flatten(), bins=256, range=[0,256], density=True)
    plt.tight_layout()
    plt.show()
```

```python
path = "C:\\Users\\Kuldeep Solanki\\Desktop\\images\\KAKASHI.jpg"
kernel = np.array([[-1,-1,-1],
                  [-1,8,-1],
                  [-1,-1,-1]])
convo(path,kernel)
```
