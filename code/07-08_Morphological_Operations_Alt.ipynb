```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("C:\\Users\\Kuldeep Solanki\\Desktop\\KAKASHI.jpg")
img1 = cv2.resize(img,(700,700))
img2 = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img1)
```

```python
img3 = img1[:,100:500]
plt.imshow(img3)
```

```python
T = np.float32([[1,0,50],[0,1,100]])
imgT = cv2.warpAffine(img1,T,(350,350))
plt.imshow(imgT)
```

```python
R = np.float32([[np.cos(np.pi/6),np.sin(np.pi/6),0],[-np.sin(np.pi/6),np.cos(np.pi/6),0]])
imgR = cv2.warpAffine(img1,R,(700,700))
plt.imshow(imgR)
```

```python
Scale = np.float32([[0.5,0,1],[0,0.5,1]])
imgS = cv2.warpAffine(img1,Scale,(700,700))
plt.imshow(imgS)
```

```python
Shearing = np.float32([[1,0.1,1],[0.05,1,1]])
imgSh = cv2.warpAffine(img1,Shearing,(700,700))
plt.imshow(imgSh)
```

```python
#image Rotation
t = cv2.getRotationMatrix2D((0,0),45,1)
print(t)
imgR1 = cv2.warpAffine(img1,t,(700,700),borderMode = cv2.BORDER_CONSTANT)
plt.imshow(imgR1)
```
