# Q-9 Perform Segmentation using
- Simple thresholding
- Otsu Binarization
- Adaptive Thresholdng


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:\\Users\\rcc\\Desktop\\KAKASHI.jpg",0)  
```

```python
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY) 
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV) 
ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC) 
ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO) 
ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)
```

```python
plt.figure(figsize=(7, 7))
plt.subplot(3, 2, 1)
plt.imshow(img,cmap='gray') 
plt.title('Original Image') 

plt.subplot(3, 2, 2)
plt.imshow(thresh1,cmap='gray') 
plt.title('Binary Threshold') 

plt.subplot(3, 2, 3)
plt.imshow(thresh2,cmap='gray') 
plt.title('Binary Threshold Inverted') 

plt.subplot(3, 2, 4)
plt.imshow(thresh3,cmap='gray') 
plt.title('Truncated Threshold') 

plt.subplot(3, 2, 5)
plt.imshow(thresh4,cmap='gray') 
plt.title('Set to 0') 

plt.subplot(3, 2, 6)
plt.imshow(thresh5,cmap='gray') 
plt.title('Set to 0 Inverted') 

plt.tight_layout()
plt.show()
```

```python
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)   
```

```python
plt.figure(figsize=(7,15))
plt.subplot(321)
plt.imshow(img,cmap='gray')
plt.subplot(322)
plt.imshow(thresh1,cmap='gray') 
```

```python
plt.figure(figsize=(7,7))
plt.subplot(321)
plt.hist(img)
plt.subplot(322)
plt.hist(thresh1)
plt.show()
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
  
thresh3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
```

```python
plt.figure(figsize=(6,10))
plt.subplot(321)
plt.imshow(img,cmap='gray')
plt.subplot(322)
plt.imshow(thresh2,cmap='gray')
plt.subplot(323)
plt.imshow(thresh3,cmap='gray')
```

```python
plt.figure(figsize=(7,7))
plt.subplot(321)
plt.hist(img.flatten())
plt.subplot(322)
plt.hist(thresh2.flatten())
plt.subplot(323)
plt.hist(thresh3.flatten())
```
