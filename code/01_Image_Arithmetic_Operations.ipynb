```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Image load and resize subplot

```python
img = cv2.imread("C:\\Users\\Kuldeep Solanki\\Desktop\\images\\doctorrrr.jpeg")
reimg = cv2.resize(img,(150,150))

plt.subplot(121)
plt.title("Original Image")
plt.axis('off')
plt.imshow(img)

plt.subplot(122)
plt.title("Resized image")
plt.axis('off')
plt.imshow(reimg)
```

```python
img2 = cv2.imread("C:\\Users\\Kuldeep Solanki\\Desktop\\images\\emol.jpeg",cv2.COLOR_RGB2BGR)
reimg2 = cv2.resize(img2,(150,150))

plt.subplot(121)
plt.title("original image")
plt.axis('off')
plt.imshow(img2)

plt.subplot(122)
plt.title("Resized image")
plt.axis('off')
plt.imshow(reimg2)
```

# ADD,SUB,MUL,DIV

```python
img_add = cv2.add(reimg,reimg2)
img_sub = cv2.subtract(reimg,reimg2)
img_mul = cv2.multiply(reimg,reimg2)
img_div = cv2.divide(reimg,reimg2)
```

```python
plt.imshow(img_add)
plt.title("Add of 2 img")
plt.axis('off')
plt.show()

plt.imshow(img_sub)
plt.title("Sub of 2 img")
plt.axis('off')
plt.show()

plt.imshow(img_mul)
plt.title("Multiply of 2 img")
plt.axis('off')
plt.show()

plt.imshow(img_div)
plt.title("Div of 2 img")
plt.axis('off')
plt.show()
```
