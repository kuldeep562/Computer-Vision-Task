```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def convo(path,kernel):
    img=cv2.imread(path,0)
    plt.figure(figsize=(7, 7))
    plt.subplot(321)
    plt.imshow(img, cmap='gray')
    plt.title('Real image')
    plt.subplot(322)
    plt.hist(img.flatten(), bins=256, range=[0,256], density=True)
    plt.title('Histogram of Real image')
    h,w=img.shape
    for i in range(h-2):
        for j in range(w-2):
            mat=np.array(img[i:i+3,j:j+3])
            ad=np.array([mat*kernel])
            ad=np.sum(ad)
            img[i,j]=abs(ad)
    plt.subplot(323)
    plt.imshow(img, cmap='gray')
    plt.title('Convolved image')
    plt.subplot(324)
    plt.hist(img.flatten(), bins=256, range=[0,256], density=True)
    plt.title('Histogram of Convolved image')
    plt.tight_layout()
    plt.show()
```

```python
path="C:\\Users\\Kuldeep Solanki\\Desktop\\images\\KAKASHI.jpg"
kernel=np.array([[-1,-1,-1],
                 [-1, 8,-1],
                 [-1,-1,-1]])
convo(path,kernel)
```
