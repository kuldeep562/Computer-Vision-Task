```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:\\Users\\Kuldeep Solanki\\Desktop\\KAKASHI.jpg')
image = cv2.resize(image,(500,500))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
def multiply_by_constant(image, constant):
    result = cv2.multiply(image, constant)
    return result

def divide_by_constant(image, constant):
    result = cv2.divide(image, constant)
    return result

def resize_image(image, scale_percent):
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    resized_image = cv2.resize(image, (width, height), interpolation=cv2.INTER_AREA)
    return resized_image

def slice_image(image, x, y, width, height):
    sliced_image = image[y:y+height, x:x+width]
    return sliced_image

def mask_image(image, mask):
    # Convert mask to binary (if it's not already)
    if len(mask.shape) > 2:
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)    
    # Resize mask to match image dimensions (if necessary)
    if image.shape[:2] != mask.shape[:2]:
        mask = cv2.resize(mask, (image.shape[1], image.shape[0]), interpolation=cv2.INTER_NEAREST)
    # Perform bitwise AND operation
    masked_image = cv2.bitwise_and(image, image, mask=mask)
    return masked_image

def add_images(image, image2):

    result = cv2.add(image, image2)
    return result

def weighted_add_images(image1, image2, alpha, beta, gamma):
    result = cv2.addWeighted(image1, alpha, image2, beta, gamma)
    return result


def subtract_images(image1, image2):
    result = cv2.subtract(image1, image2)
    return result

def logical_operations(image):
    # Creating a binary mask
    mask = cv2.inRange(image, (100, 100, 100), (255, 255, 255))    
    bitwise_and = cv2.bitwise_and(image, image, mask=mask)
    bitwise_or = cv2.bitwise_or(image, image, mask=mask)
    bitwise_not = cv2.bitwise_not(image, mask=mask)
    return bitwise_and, bitwise_or, bitwise_not
```

```python
constant = 2
resized_image = resize_image(image, 50)
sliced_image = slice_image(image, 100, 100, 200, 200)
mask = cv2.imread('C:\\Users\\rcc\\Desktop\\download.jpg', cv2.IMREAD_GRAYSCALE)
masked_image = mask_image(image, mask)
image2 = cv2.imread('C:\\Users\\rcc\\Desktop\\images.jpg ')
image2 = cv2.resize(image2,(500,500))
added_image = add_images(image, image2)
weighted_added_image = weighted_add_images(image, image2, 0.7, 0.3, 0)
subtracted_image = subtract_images(image, image2)
bitwise_and, bitwise_or, bitwise_not = logical_operations(image)
```
